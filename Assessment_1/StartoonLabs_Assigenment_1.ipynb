```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Read data from files
data_1 = np.loadtxt('/content/Data_1.txt')

# Assuming data_1 contains the signal values
min_dis = 40
# Find peaks and troughs
peaks= find_peaks(data_1,min_dis)[0]
troughs = find_peaks(-data_1,distance=min_dis)[0]


# Plot the signal
plt.figure(figsize=(10,5))
plt.plot(data_1, label='Signal')
plt.plot(peaks, data_1[peaks], 'ro', label='Maxima')  # Red dots for maxima
plt.plot(troughs, data_1[troughs], 'bo', label='Minima')  # Blue dots for minima
plt.xlabel('Index')
plt.ylabel('Signal Value')
plt.title('Signal with Peaks and Troughs')
plt.legend()

# Save the plot
plt.savefig('signal_plot.png')

# Show the plot
plt.show()

# Print peaks and troughs
print("Peaks indices:", peaks)
print("Troughs indices:", troughs)
```
